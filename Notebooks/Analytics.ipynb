Advanced Analytics Summary

Key Insights

1. Historical VaR & CVaR : 
The Historical VaR and CVaR analysis identified the funds with the highest downside risk during adverse market conditions. Funds with lower VaR values were comparatively more stable.

2. Rolling Sharpe Ratio :
The Rolling 90-Day Sharpe Ratio demonstrated that risk-adjusted performance changes over time. Funds with consistently higher rolling Sharpe values delivered superior performance.

3. Investor Cohort Analysis : 
The cohort analysis highlighted the investment behaviour of investors based on their first investment year. The cohort with the highest average SIP amount contributed the largest investment.

4. SIP Continuation Analysis : 
Investors with an average SIP gap greater than 35 days were classified as "At Risk", indicating a higher probability of SIP discontinuation.


5. Sector Concentration Risk : 
HHI analysis revealed the diversification level of each mutual fund. Funds with lower HHI values were more diversified, whereas funds with higher HHI values showed greater sector concentration.
